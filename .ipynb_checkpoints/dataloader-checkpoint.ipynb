```python
import numpy as np
```

```python

# Read binary file
with open('A-EOG_raw_data_example.bin', "rb") as f:
    byte_data = f.read()  # Read the entire file as bytes


# Function to swap adjacent byte pairs
def swap_adjacent_pairs(hex_str):
    return "".join([hex_str[i:i+2] for i in range(0, len(hex_str), 2)][::-1])
    
def locate_first_a(hex_string_ori):
    index = hex_string_ori.find('aaaaaaaa')
    print(f'We found index of the first aaaaaaaa{index}')
    index = hex_string_ori.find('aaaaaaaa', index+1)
    print(f'We found index of the second aaaaaaaa{index}')
    index = index - 7*8
    print(f'The index {index} is the start of the second chunk of 16 channels')
    return index

def discard_redundant(chunks):
    if chunks and len(chunks[-1]) < 8:
        chunks.pop()  # Remove the last element   
    return chunks

    
# Convert bytes to a hexadecimal string
hex_string_ori = byte_data.hex()
print('----raw data')
# print(hex_string_ori[:28*8])
index = locate_first_a(hex_string_ori)
print(hex_string_ori[index:28*8])
hex_string = hex_string_ori[index:]


chunk_size = 8  # 8 hex characters = 32 bits

ori_chunk = [hex_string[i:i+chunk_size] for i in range(0, len(hex_string), chunk_size)]
print('----Original chunk:')
print(ori_chunk[:14])

# Apply the function to each element in the list
chunks = [swap_adjacent_pairs(item) for item in ori_chunk]
print('----Reversed order of each chunk:')
print(chunks[:20])
discard_redundant(chunks)

print(len(chunks))
print(type(chunks))

# synchronization
# 1. detect 8 digits of a --- done
# 2. locate the beginning of the string --- done
# 3. reorder the every two digits in each chunk element to find the timestamp (8 digits before aaaaaaaa)

```

```python
def hex_to_signed_int(hex_str):
    int_value = int(hex_str, 16)
    if int_value >= 2**23:  # If the value is negative in twoâ€™s complement
        int_value -= 2**24
    return int_value



def info_from_digit(chunks):
    ch_chunks = []
    vol_chunks = []
    timestamp_chunks = []
    
    count=0
    while len(chunks)%14 != 0:
        chunks.pop()
        
    for idx in range(42*5+42*5, 42*5+42*5+42*5, 14):
        chunk_element = chunks[idx:idx+14]
        print(chunk_element)
        timestamp = hex_to_signed_int(chunk_element[6]) 
        for e_idx, chunk_e in enumerate(chunk_element):
            timestamp_chunks.append(timestamp) 
            
            if e_idx ==6 or e_idx ==7:
                ch = e_idx
                ch_chunks.append(ch)
                signed_vol = np.nan
                vol_chunks.append(signed_vol)
            else:
                # vol
                ch = int(chunk_e[:2], 16)
                ch_chunks.append(ch)
                signed_int = hex_to_signed_int(chunk_e[2:])  # Example: -1
                # print(signed_int)  # Output: -1
                signed_vol = signed_int* 5.00 / (2**(24-1)-1)
                vol_chunks.append(signed_vol)

    return vol_chunks, ch_chunks, timestamp_chunks
                

vol_chunks, ch_chunks, timestamp_chunks = info_from_digit(chunks)           
        
    


    

# chunks[85: 85+14]

# print(timestamp_chunks[:28*4])
# print(vol_chunks[:28*4])
# print(ch_chunks[:28*4])

```
